### Line Plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.random.randn(100)
```

```python
plt.plot(y)
plt.autoscale(tight='x')
plt.tight_layout()
plt.show()
```

```python
plt.plot(y, 'g--')
plt.autoscale(tight='x')
plt.tight_layout()
plt.show()
```

```python
x = np.cumsum(np.random.rand(100))
```

```python
plt.plot(x,y,'r')
plt.autoscale(tight='x')
plt.tight_layout()
plt.show()
```

```python
plt.plot(x,y,alpha = 0.5, color = '#FF7F00',
         label = 'Line Label', linestyle = '-.',
         linewidth = 1, marker = 'o', markeredgecolor = '#000000',
         markeredgewidth = 2, markerfacecolor = '#FF7F00',
         markersize=5)
plt.tight_layout()
plt.show()
```

```python
y = np.random.randn(10)
```

```python
h = plt.plot(y)
plt.tight_layout()
plt.show()
```

```python
plt.getp(h)
```

```python
plt.setp(h, 'alpha')
```

```python
plt.setp(h, 'color')
```

```python
plt.setp(h, 'linestyle')
```

```python
plt.setp(h, 'linestyle', 'dashed') # Change the line style
```

```python
h = plt.plot(y)
plt.setp(h, 'linewidth', 1)
plt.setp(h, 'marker', 'o')
plt.setp(h, 'markeredgecolor', '#000000')
plt.setp(h, 'markeredgewidth', '2')
plt.setp(h, 'markerfacecolor', '#FF7F00')
plt.setp(h, 'markersize', 5)
plt.setp(h, 'color', '#FF7F00') 
plt.setp(h, 'linestyle', 'dashed')
plt.tight_layout()
plt.show()
```
