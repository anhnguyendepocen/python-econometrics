### Multiple Plots

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)

# Add the subplot to the figure
# Panel 1
y = np.random.randn(100)
ax1.plot(y)
ax1.set_title('line plot')

# Panel 2
y = np.random.rand(5)
x = np.arange(5)
ax2.bar(x, y)
ax2.set_title('bar chart')

# Panel 3
y = np.random.rand(5)
y = y / np.sum(y)
y[y < .05] = .05
colors = sns.dark_palette("skyblue", 8, reverse = True)
ax3.pie(y, colors = colors)
ax3.axis("equal")
ax3.set_title('pie chart')

# Panel 4
z = np.random.randn(100, 2)
z[:, 1] = 0.5 * z[:, 0] + np.sqrt(0.5) * z[:, 1]
x = z[:, 0]
y = z[:, 1]
ax4.scatter(x, y)
ax4.set_title('scatter plot')

plt.tight_layout()
```

```python
fig, ax = plt.subplots()

x = np.random.randn(1000)
ax.hist(x, bins=30, label='Empirical')

xlim = ax.get_xlim()
ylim = ax.get_ylim()
pdfx = np.linspace(xlim[0], xlim[1], 200)
pdfy = stats.norm.pdf(pdfx)
pdfy = pdfy / pdfy.max() * ylim[1]
ax.plot(pdfx, pdfy, 'r', label='PDF')

plt.legend()
plt.tight_layout()
```

```python
x = np.cumsum(np.random.randn(100,3), axis = 0)
plt.plot(x[:,0],'b-', label = 'Series 1')
plt.plot(x[:,1],'g-', label = 'Series 2')
plt.plot(x[:,2],'r:', label = 'Series 3')
plt.legend()
plt.title('Basic Legend')
plt.tight_layout()
```
