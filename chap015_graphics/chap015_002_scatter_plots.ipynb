### Scatter Plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.random.randn(100,2)
z[:,1] = 0.5*z[:,0] + np.sqrt(0.5)*z[:,1]
x = z[:,0]
y = z[:,1]
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y, s = 60, c = '#FF7F00', marker = 's',
            alpha = .5, label = 'Scatter Data')
plt.show()
```

```python
size_data = np.exp(np.exp(np.exp(np.random.rand(100))))
size_data = 200 * size_data / np.amax(size_data)
size_data[size_data < 1] = 1.0
plt.scatter(x,y, s = size_data, c = '#FF7F00', marker='s',
            label = 'Scatter Data')
plt.show()
```
